### 1. Daily Cost & Requests by Org and Service (Range Query)
**Requirement**: Daily costs and requests by organization and service within a date range.

**Collection**: `org_daily_usage_by_service`

```python
import pandas as pd

# -----------------------------
# Query Parameters
# -----------------------------
ORG_ID = "org_253j2d54"
SERVICE = "COMPUTE"
START_DATE = "2025-01-01"
END_DATE = "2025-12-31"

cql_q1 = f"""
-- Q1 (CQL) - Daily usage for one org + service in a date range
SELECT date, service, total_cost_usd, total_requests
FROM org_daily_usage_by_service
WHERE org_id = '{ORG_ID}'
  AND service = '{SERVICE}'
  AND date >= '{START_DATE}'
  AND date <= '{END_DATE}';
""".strip()

print(cql_q1)

usage_df = pd.DataFrame()

col_usage = db.get_collection("org_daily_usage_by_service")

query_filter = {
    "org_id": ORG_ID,
    "service": SERVICE,
    "date": {"$gte": START_DATE, "$lte": END_DATE}
}

projection = {
    "date": 1,
    "service": 1,
    "total_cost_usd": 1,
    "total_requests": 1,
    "_id": 0
}

cursor = col_usage.find(
    filter=query_filter,
    projection=projection
)

usage_df = pd.DataFrame(list(cursor))

usage_df
```

```python
from datetime import datetime, timedelta, timezone
import pandas as pd

# -----------------------------
# Query Parameters
# -----------------------------
ORG_ID = "org_253j2d54"
N = 3
TIME_WINDOWS = [14, 30, 60, 90, 120, 150, 180, 210]

END_DATE = datetime.now(timezone.utc)
start_date = datetime.now(timezone.utc) - timedelta(days=30)

cql_q2 = f"""
-- Q2 (CQL) - Fetch raw rows for a window (aggregation TOP-N is client-side)
-- Window: last 30 days ({start_date} -> {END_DATE})
SELECT service, total_cost_usd
FROM org_daily_usage_by_service
WHERE org_id = '{ORG_ID}'
  AND date >= '{start_date.isoformat()}'
  AND date <= '{END_DATE.isoformat()}';
""".strip()

print(cql_q2)

top_services_df = pd.DataFrame()
used_window = None

if "db" in locals():
    col_usage = db.get_collection("org_daily_usage_by_service")

    end_date_str = END_DATE.date().isoformat()

    for window in TIME_WINDOWS:
        start_date = END_DATE - timedelta(days=window)

        query_filter = {
            "org_id": ORG_ID,
            "date": {
                "$gte": start_date.date().isoformat(),
                "$lte": end_date_str
            }
        }

        projection = {
            "service": 1,
            "total_cost_usd": 1,
            "_id": 0
        }

        cursor = col_usage.find(
            filter=query_filter,
            projection=projection
        )

        docs = list(cursor)

        if docs:
            df = pd.DataFrame(docs)
            used_window = window
            break

    # -----------------------------
    # Client-side aggregation
    # -----------------------------
    if not df.empty:
        top_services_df = (
            df.groupby("service", as_index=False)["total_cost_usd"]
              .sum()
              .sort_values("total_cost_usd", ascending=False)
              .head(N)
        )

        top_services_df["window_days"] = used_window

top_services_df
```

### 3. Tickets Evolution & SLA Breach Rate
**Requirement**: Evolution of critical tickets and SLA breach rate per day.

**Collection**: `tickets_by_org_date`

```python
TIME_WINDOWS = [30, 210] # Try 30 days first

col_tickets = db.get_collection("tickets_by_org_date")

# ------------------------------------------------------------------
# Extracción con fallback temporal
# ------------------------------------------------------------------
df = pd.DataFrame()

today = datetime.utcnow().date()
start_date = today - timedelta(days=TIME_WINDOWS[0])

cql_q3 = f"""
-- Q3 (CQL) - Fetch CRITICAL tickets in a date window
-- NOTE: Cassandra typically requires partition key filters (e.g., org_id).
-- This query assumes a table/model that supports querying by severity + date.
SELECT ticket_date, severity, total_tickets, sla_breach_count, avg_resolution_hours
FROM tickets_by_org_date
WHERE severity = 'critical'
  AND ticket_date >= '{start_date.isoformat()}'
  AND ticket_date <= '{today.isoformat()}';
""".strip()

print(cql_q3)

for days in TIME_WINDOWS:
    start_date = today - timedelta(days=days)

    print(f"Trying last {days} days (from {start_date})...")

    docs = col_tickets.find(
        filter={
            "severity": "critical"
        },
        projection={
            "ticket_date": 1,
            "severity": 1,
            "total_tickets": 1,
            "sla_breach_count": 1,
            "avg_resolution_hours": 1,
            "_id": 0
        }
    )

    tmp_df = pd.DataFrame(list(docs))

    if tmp_df.empty:
        continue

    # Asegurar tipo fecha y filtrar por ventana
    tmp_df["ticket_date"] = pd.to_datetime(tmp_df["ticket_date"]).dt.date
    tmp_df = tmp_df[tmp_df["ticket_date"] >= start_date]

    if not tmp_df.empty:
        df = tmp_df
        print(f"Data found for last {days} days")
        break

# ------------------------------------------------------------------
# Si no hay datos
# ------------------------------------------------------------------
if df.empty:
    print("No data found up to 210 days")
    result_df = df

else:
    # ------------------------------------------------------------------
    # Re-agregación global (sin org_id)
    # ------------------------------------------------------------------
    result_df = (
        df.groupby(["ticket_date", "severity"], as_index=False)
          .apply(
              lambda g: pd.Series({
                  "Total Tickets": g["total_tickets"].sum(),
                  "SLA Breach Rate": (
                      g["sla_breach_count"].sum() / g["total_tickets"].sum()
                      if g["total_tickets"].sum() > 0 else 0.0
                  ),
                  "Avg Resolution Hours": (
                      (g["avg_resolution_hours"] * g["total_tickets"]).sum()
                      / g["total_tickets"].sum()
                      if g["total_tickets"].sum() > 0 else None
                  )
              })
          )
          .reset_index(drop=True)
    )

    result_df = result_df.rename(columns={
        "ticket_date": "Date",
        "severity": "Severity"
    })

    result_df["Date"] = pd.to_datetime(result_df["Date"])
    result_df = result_df.sort_values("Date")

# ------------------------------------------------------------------
# Resultado final
# ------------------------------------------------------------------
result_df
```

### 4. Monthly Revenue
**Requirement**: Monthly revenue, ordered by latest date.

**Collection**: `revenue_by_org_month`

```python
import pandas as pd

# -----------------------------
# Query Parameters
# -----------------------------
ORG_ID = "org_253j2d54"

revenue_df = pd.DataFrame()

cql_q4 = f"""
-- Q4 (CQL) - Last 12 months revenue for one org
SELECT year, month, total_revenue_usd
FROM revenue_by_org_month
WHERE org_id = '{ORG_ID}'
ORDER BY year DESC, month DESC
LIMIT 12;
""".strip()

print(cql_q4)

if "db" in locals():
    col_revenue = db.get_collection("revenue_by_org_month")

    query_filter = {"org_id": ORG_ID}
    sort = {"year": -1, "month": -1}

    cursor = col_revenue.find(
        filter=query_filter,
        sort=sort,
        limit=12,
        projection={
            "year": 1,
            "month": 1,
            "total_revenue_usd": 1,
            "_id": 0
        }
    )

    revenue_df = pd.DataFrame(list(cursor))

    if not revenue_df.empty:
        revenue_df["period"] = pd.to_datetime(
            revenue_df["year"].astype(str) + "-" +
            revenue_df["month"].astype(str) + "-01"
        )

        revenue_df = revenue_df.sort_values("period")

revenue_df
```

### 5. GenAI Tokens & Cost
**Requirement**: Daily GenAI usage stats.

**Collection**: `genai_tokens_by_org_date`

```python
ORG_ID = "org_253j2d54"

cql_q5 = f"""
-- Q5 (CQL) - GenAI tokens and cost by day for one org
SELECT org_id, year, month, day, total_tokens, genai_cost_usd
FROM genai_tokens_by_org_date
WHERE org_id = '{ORG_ID}'
ORDER BY year ASC, month ASC, day ASC;
""".strip()

print(cql_q5)

col_genai = db.get_collection("genai_tokens_by_org_date")

docs = list(
    col_genai.find(
        filter={"org_id": ORG_ID},
        projection={
            "org_id": 1,
            "year": 1,
            "month": 1,
            "day": 1,
            "total_tokens": 1,
            "genai_cost_usd": 1
        }
    )
)

df = pd.DataFrame(docs)

# Limpieza defensiva
df = df.drop(columns=["_id"], errors="ignore")

# Verificación explícita (útil para debug)
print(df.columns)

# Ordenar
df = df.sort_values(["year", "month", "day"]).reset_index(drop=True)

df
```
